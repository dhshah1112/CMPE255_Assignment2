```python
from pycaret.clustering import *
clustering_setup = setup(df, use_gpu = True)
```

```python
kmeans = create_model('kmeans')
```

```python
kmeans = create_model('kmeans', num_clusters= 5)
```

```python
# Plot a cluster
plot_model(kmeans)

# Plot a specific detail like distribution of a feature in each cluster
plot_model(kmeans, plot='distribution')  # replace 'distribution' with other available plots
```
