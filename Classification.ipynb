```python
df_1 = pd.read_csv('/content/drive/MyDrive/CMPE255/winequality-red.csv')
```

```python
from pycaret.classification import *
setup(df_1, target='quality')
```

```python
compare_models()
```

```python
model = create_model('et', cross_validation=True)
```

```python
tuned_model = tune_model(model)
```

```python
ensemble_model = ensemble_model(tuned_model)
```

```python
evaluate_model(ensemble_model)
```

```python
import gradio as gr

# Adjusted prediction function to use 'prediction_label'
def predict_quality(*args):
    input_data = pd.DataFrame([args], columns=df_1.columns[:-1])  # excluding the target column
    prediction = predict_model(model, data=input_data)
    return prediction['prediction_label'].iloc[0]

# Define the Gradio interface
features = [gr.inputs.Number(label=column) for column in df_1.columns[:-1]]
iface = gr.Interface(fn=predict_quality,
                     inputs=features,
                     outputs=gr.outputs.Textbox(label="Predicted Quality"))

# Run the Gradio app
iface.launch(share=True)
```
